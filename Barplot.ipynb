```python
import matplotlib.pyplot as plt
```

```python
gender=['Male', 'Female']
freq=[14, 18]
plt.bar(x=gender, height=freq)
plt.xlabel("gender")
plt.ylabel("Count/Frequency")
plt.show()
```

```python
import seaborn as sns
gender=['Male', 'Female']
freq=[14, 18]
sns.barplot(x=gender, y=freq)
plt.xlabel("gender")
plt.ylabel("Count/Frequency")
plt.show()
```
