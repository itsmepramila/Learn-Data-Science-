# kurtosis
kurtosis is the degree of presence of outliers in the distribution
* mesokurtic: Normal distribution, kurtosis=3(pearson formula)
* platykurtic: Uniform distribution , kurtosis<3(pearson formula)
* leptokurtic: eg, laplace distibution, kurtosis>3(pearson formula)

# Mesokurtic

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(0)
data=np.random.normal(loc=5, scale=1, size=15000)
```

```python
from scipy.stats import kurtosis
#If true, fisher definition is used (normal==0.0). if false , pearson pearson definition is used(normal==>3.0)
kurt = kurtosis(data, fisher=False)
```

```python
sns.histplot(data, kde=True)
plt.title(f"kurtosis={round(kurt, 2)}")
```

```python
# outlier frequency is medium
```

# platykurtic
* no outliers

```python
np.random.seed(0)
data=np.random.uniform(low=2, high=5, size=15000)
kurt=kurtosis(data, fisher=False)
sns.histplot(data, kde=True, bins=3)
plt.title(f"kurtosis={round(kurt, 2)}")
plt.show()
```

```python
np.random.seed(0)
data=np.random.uniform(low=5, high=1000, size=15000)
kurt=kurtosis(data, fisher=False)
sns.histplot(data, kde=True, bins=3)
plt.title(f"kurtosis={round(kurt, 2)}")
plt.show()
```

# Leptokutic
* extreme outlier

```python
np.random.seed(0)
data=np.random.laplace(loc=5, scale=1, size=15000)
kurt=kurtosis(data, fisher=False)
sns.histplot(data, kde=True)
plt.title(f"kurtosis={round(kurt, 2)}")
plt.show()
```
