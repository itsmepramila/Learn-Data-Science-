```python
import matplotlib.pyplot as plt
```

# linechart

```python
subject=['Math', 'English', 'Science', 'Opt']
marks=['40', '50', '70', 80]
plt.plot(subject, marks)
```

```python
subject=['Math', 'English', 'Science', 'Opt']
marks=['40', '50', '70', 80]
plt.plot(subject, marks, 'o-')
```

```python
plt.plot(subject, marks, 'o-', color='purple', ms=10)
```

```python
plt.plot(subject, marks, 'o-')
plt.xlabel("Subeject")
plt.ylabel("Marks")
plt.show()
```

```python
plt.plot(subject, marks, 'o-')
plt.xlabel("Subeject")
plt.ylabel("Marks")
plt.title("Marks in Different Subject")
plt.show()
```
