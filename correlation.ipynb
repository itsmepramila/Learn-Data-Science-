```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
np.random.seed(200)
x=np.random.randn(200)
y=x*10+np.random.randn(200)*2
```

```python
res=stats.pearsonr(x,y)
cor, p_val=res
print(f"pearson correlation coef. {round(cor, 3)}")
```

```python
plt.scatter(x=x, y=y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
