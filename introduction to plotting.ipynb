# line chart

```python
import matplotlib.pyplot as plt
```

```python
month=['Jan', 'Feb', 'Mar', 'Apr', 'May']
growth=[3.2,5,6,5.5,7]
```

```python
plt.plot(month, growth)
```

```python
plt.plot(month, growth, 'o-') # o dot show garna use garinxa # o not 0
```

```python
plt.plot(month, growth, 'o-', color='purple')
```

```python
plt.plot(month, growth, 'o-', color='purple', ms=10) # ms means marker-size
```

```python
plt.plot(month, growth, 'o-', color='purple', ms=10)
plt.xlabel("Months") # x and y label means x-axis and y-axis title
plt.ylabel("Growth (%)")
```

```python
plt.plot(month, growth, 'o-', color='purple', ms=10)
plt.xlabel("Months") # x and y label means x-axis and y-axis title
plt.ylabel("Growth (%)")
plt.show() # plt.show le text hataune kaam garxa
```

```python
plt.plot(month, growth, 'o-', color='purple', ms=10)
plt.xlabel("Months") # x and y label means x-axis and y-axis title
plt.ylabel("Growth (%)")
plt.title("Monthly Growth (%)") # plt.title gives the title of the linechart
plt.show() # plt.show le text hataune kaam garxa
```

```python
month=['Jan', 'Feb', 'Mar', 'Apr', 'May']
actual_growth=[3.2, 5, 6, 5.5, 7 ]
predicted_growth=[3.5,6,6.5,6,8]
plt.plot(month, actual_growth, 'o-')
plt.plot(month, predicted_growth, 'o-')
plt.xlabel("Months")
plt.ylabel('Growth (%)')
plt.show()
```

```python
month=['Jan', 'Feb', 'Mar', 'Apr', 'May']
actual_growth=[3.2, 5, 6, 5.5, 7 ]
predicted_growth=[3.5,6,6.5,6,8]
plt.plot(month, actual_growth, 'o-', label='Actual Growth', color='red')
plt.plot(month, predicted_growth, 'o-', label='Predicted Growth', color='purple')
plt.xlabel("Months")
plt.ylabel('Growth (%)')
plt.legend() # plt.legend le label plt.plot ko label parameter laii plot garxa
plt.show()
```

```python
month=['Jan', 'Feb', 'Mar', 'Apr', 'May']
actual_growth=[3.2, 5, 6, 5.5, 7 ]
predicted_growth=[3.5,6,6.5,6,8]
plt.plot(month, actual_growth, 'o-', label='Actual Growth', color='red', linewidth=2) # linewidth minimize the width of line
plt.plot(month, predicted_growth, 'o-', label='Predicted Growth', color='purple', linewidth=1)
plt.xlabel("Months")
plt.ylabel('Growth (%)')
plt.legend() # plt.legend le label plt.plot ko label parameter laii plot garxa
plt.show()
```

```python
month=['Jan', 'Feb', 'Mar', 'Apr', 'May']
actual_growth=[3.2, 5, 6, 5.5, 7 ]
predicted_growth=[3.5,6,6.5,6,8]
plt.plot(month, actual_growth, 'o-', label='Actual Growth', color='red', linewidth=2) # linewidth minimize the width of line
plt.plot(month, predicted_growth, 'o-', label='Predicted Growth', color='purple', linewidth=1)
plt.xlabel("Months")
plt.ylabel('Growth (%)')
plt.legend(loc='lower right') # loc='lower right' lower ma right ma rakhne kaam garxa
plt.show()
```

# barplot

```python
gender=['Male', 'Female']
freq=[14, 18]
plt.bar(x=gender, height=freq)
plt.show()
```

```python
gender=['Male', 'Female']
freq=[14, 18]
plt.bar(x=gender, height=freq)
plt.xlabel("Gender")
plt.ylabel("Count/Frequency")
plt.show()
```

```python
import seaborn as sns
gender=['Male', 'Female']
freq=[14, 18]
sns.barplot(x=gender, y=freq)
plt.xlabel("Gender")
plt.ylabel("Count/Frequency")
plt.show()
```
