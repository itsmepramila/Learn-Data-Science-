```python
import pandas as pd
df = pd.read_csv('old_faithful.csv')
```

```python
df.head()
```

```python
import seaborn as sns
sns.scatterplot(x=df['eruptions'], y=df['waiting'])
```

```python
no_of_clusters=range(1,7)
```

```python
from sklearn.cluster import KMeans
sse=[]
for k in no_of_clusters:
    model=KMeans(n_clusters=k, n_init='auto', random_state=40)
    model.fit(df)
    sse.append(model.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(no_of_clusters, sse, 'o-', color='purple', ms=10)
plt.xlabel("No of Clusters(K)")
plt.ylabel("Sum of squared distances")
plt.show()
```

The elbow mwthos show the optimal number of cluster is 2

```python
k=2
model=KMeans(n_clusters=k, n_init='auto', random_state=40)
clusters=model.fit_predict(df)
```

```python
import seaborn as sns
sns.scatterplot(x=df['eruptions'], y=df['waiting'], hue=clusters)
sns.scatterplot(x=model.cluster_centers_[:, 0], y=model.cluster_centers_[:, 1], marker='*', s=500, color=['red', 'green'])
```
