import matplotlib.pyplot as plt

# linechart

```python
subject=['Math', 'English', 'Science', 'Opt']
marks=['40', '50', '70', 80]
plt.plot(subject, marks)
```

```python
subject=['Math', 'English', 'Science', 'Opt']
marks=['40', '50', '70', 80]
plt.plot(subject, marks, 'o-')
```

```python
plt.plot(subject, marks, 'o-', color='purple', ms=10)
```

```python
plt.plot(subject, marks, 'o-')
plt.xlabel("Subeject")
plt.ylabel("Marks")
plt.show()
```

```python
plt.plot(subject, marks, 'o-')
plt.xlabel("Subeject")
plt.ylabel("Marks")
plt.title("Marks in Different Subject")
plt.show()
```

# sactterplot

```python
import matplotlib.pyplot as plt

day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
colors = ['r', 'y', 'g', 'b', 'r', 'g', 'r']
sizes=[400, 200, 400, 300, 200, 100, 600]
plt.scatter(day, no, c=colors, s=sizes, alpha=1, marker="*", edgecolor='g', linewidth=2)
plt.title("Scatterplot")
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
colors = ['r', 'y', 'g', 'b', 'r', 'g', 'r']
sizes=[400, 200, 400, 300, 200, 100, 600]
plt.scatter(day, no, c=colors, s=sizes)
plt.title("Scatterplot")
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```
