````python
# i want row 11, 12, 13, 14, 15 and all columns
df.loc[11:15, :]
````

# iloc
```python
df.iloc[row_index, column_index]
```
* iloc ma column index ko thauma column name rakhana paudina

````python
df.iloc[12:15, 2:] # row numbwr 12, 13,14 liney rw 2 paxadiko sabaii line
````

````python
df.iloc[:5, 1:4] # first five rows, and column index 1,2, 4
````

````python
df.loc[14:16, ['size', 'sex']]
````

````python
df.iloc[14:17, -3:]
````

````python
df.iloc[14:17, -4: -2]
````

# groupby

````python
df.groupby(['time']).tip.mean()
````

````python
# on average kun day badi tip dinxa
df.groupby(['day']).tip.mean()
````

````python
df.groupby(['sex']).tip.mean()
````

````python
# male smoker le badi tip ki male non_smoker on average
# female smokr le badi ki female non_smoker le on average
df.groupby(['sex', 'smoker']).tip.mean()
````

````python
df.groupby(['size']).tip.mean()
````

# Loc and iloc are two functions in Pandas that are used to slice a data set in a Pandas DataFrame. The function . loc is typically used for label indexing and can access multiple columns, while . iloc is used for integer indexing.
